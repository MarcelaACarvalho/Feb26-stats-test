# Task 1) Calculate the expected value for the club for both games.

```python
import random
import string
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
import statistics as st
```

**GAME 1**

```python
# The player pays €5 and picks 3 letters (all lower case English letters) and 2 digits (numbers 0-9 inclusive). They win as follows:

    # What is correct            Prize
    # 3 letters, both numbers	  €100
    # Two letters, both numbers	  €50
    # One letter, both numbers	  €25
    # No letter, both numbers	  €10

# Matched x Approximate Odds

all_letters = 1/26 * 1/26 * 1/26 * 1/10 * 1/10
print(all_letters)
two_letters = 1/26 * 1/26 * 25/26 * 1/10 * 1/10
print(two_letters)
one_letters = 1/26 * 25/25 * 25/26 * 1/10 * 1/10
print(one_letters)
no_letters = 25/26 * 25/25 * 25/26 * 1/10 * 1/10
print(no_letters)
```

```python
jackpot = 100 #This is the maximum amount a person can win in this game.
win_data = [
    [jackpot, all_letters],
    [50, two_letters],
    [25, one_letters],
    [10, no_letters]
]

expected_value_game01 = 0
for wins in win_data:
    expected_value_game01 += wins[0] * wins[1]
    print(f"Ex: {wins[0] * wins[1]}")
print("expected_value_player_game1:", expected_value_game01)
print("expected_value_club_game1:",5 - expected_value_game01) # 5 represents the amount paid for each play.
```

**Expected value** means how much the player or club will "receive" on average for each play.
In this case the player will "receive" €0.10 euros and the club will "receive" €4.90 euros.

On average, you lose about €4.9 per game and win €0.10
Imagine playing 10,000 times:

You would spend:
→ €5 × 10,000 = €50,000

You would win (on average):
→ €0.12 × 10,000 = €1,200

For the club, the thinking is exactly the opposite:
It earns €4.90 for each ticket

**GAME 02**

The player pays €5 and picks 3 letters (all lower case AND upper case English letters, repetition allowed). 
If they guess all correctly, they win €500,000

```python
# Matched x Approximate Odds

letters = 1/52 * 1/52 * 1/52
print(letters)

jackpot = 500_000
expected_value_game02 = jackpot * letters

print("expected_value_player:", expected_value_game02)
print("expected_value_club:", 5 - expected_value_game02)
```

On average, you lose about €1.44 per game and win €3.55
Imagine playing 10,000 times:

You would spend:
→ €5 × 10,000 = €50,000

You would win (on average):
→ €3.55 × 10,000 = €3,550

For the club, the thinking is exactly the opposite:
It wins €1.44 for each tickets

In this game, the player has a higher average gain than in the first game.

# Task 2) Create simulations of each game in Python

**SIMULATION - GAME 1**

```python
# Simulating a random event using a probability value between 0 and 1
# This value will determine whether the letter is changed or kept

def spin_letter():
    outcome = random.random()
    
    if outcome < 1/26:
        return "a"
    elif outcome < 2/26:
        return "b"
    elif outcome < 3/26:
        return "c"
    elif outcome < 4/26:
        return "d"
    elif outcome < 5/26:
        return "e"
    elif outcome < 6/26:
        return "f"
    elif outcome < 7/26:
        return "g"
    elif outcome < 8/26:
        return "h"
    elif outcome < 9/26:
        return "i"
    elif outcome < 10/26:
        return "j"
    elif outcome < 11/26:
        return "k"
    elif outcome < 12/26:
        return "l"
    elif outcome < 13/26:
        return "m"
    elif outcome < 14/26:
        return "n"
    elif outcome < 15/26:
        return "o"
    elif outcome < 16/26:
        return "p"
    elif outcome < 17/26:
        return "q"
    elif outcome < 18/26:
        return "r"
    elif outcome < 19/26:
        return "s"
    elif outcome < 20/26:
        return "t"
    elif outcome < 21/26:
        return "u"
    elif outcome < 22/26:
        return "v"
    elif outcome < 23/26:
        return "w"
    elif outcome < 24/26:
        return "x"
    elif outcome < 25/26:
        return "y"
    else:
        return "z"
```

```python
# Executing the spin_letter function to simulate a random letter outcome
spin_letter()
```

```python
# This value will determine whether the number is changed or kept

def spin_number():
    outcome = random.random()
    
    if outcome < 1/10:
        return 0
    elif outcome < 2/10:
        return 1
    elif outcome < 3/10:
        return 2
    elif outcome < 4/10:
        return 3
    elif outcome < 5/10:
        return 4
    elif outcome < 6/10:
        return 5
    elif outcome < 7/10:
        return 6
    elif outcome < 8/10:
        return 7
    elif outcome < 9/10:
        return 8
    else:
        return 9
```

```python
# Executing the spin_letter function to simulate a random number outcome
spin_number()
```

```python
# Game Simulation 1: simulating the lottery game to evaluate outcomes and prize distribution

# Defining the player's chosen letters and numbers
player_letters = ["a", "b", "c"]
player_numbers = [1, 2]

def game01():
    
    # Generating random letters
    letter1 = spin_letter()
    letter2 = spin_letter()
    letter3 = spin_letter()
    
    # Generating random numbers
    number1 = spin_number()
    number2 = spin_number()
    
    # Storing the generated results
    outcome_letters = [letter1, letter2, letter3]
    outcome_numbers = [number1, number2]

    # Initializing counters for correct matches
    correct_letters = 0
    correct_numbers = 0

    # Counting how many player letters match the outcome
    for letter in player_letters:
        if letter in outcome_letters:
            correct_letters += 1

    # Counting how many player numbers match the outcome
    for number in player_numbers:
        if number in outcome_numbers:
            correct_numbers += 1

    # Assigning prize based on matching rules
    if correct_numbers == 2 and correct_letters == 3:
        prize = 100
    elif correct_numbers == 2 and correct_letters == 2:
        prize = 50
    elif correct_numbers == 2 and correct_letters == 1:
        prize = 25
    elif correct_numbers == 2 and correct_letters == 0:
        prize = 10
    else:
        prize = 0

    # Combining letters and numbers into a single outcome
    outcome = outcome_letters + outcome_numbers

    # Returning the final result (outcome and prize)
    return outcome, prize
```

```python
# Running the first game simulation
game01()
```

```python
# Defining the player's chosen letters (case-sensitive)
player_letters = ["a", "B", "c"]

# Simulating Game 2 with case-sensitive letter matching
def game2():
    
    # Generating three random letters using the spin function
    letter1 = spin_letter2()
    letter2 = spin_letter2()
    letter3 = spin_letter2()

    # Storing the generated outcome as a list
    outcome = [letter1, letter2, letter3]

    # Checking if the generated outcome matches the player's letters exactly
    if outcome == player_letters:
        prize = 500000  # Jackpot prize
        
    else:
        prize = 0  # No prize

    # Returning the result of the simulation (outcome and prize)
    return outcome, prize
```

In this version of the game, uppercase and lowercase letters are treated as different values,

# Task 3) Calculate the measures of central tendency for each game, based off an appropriate number of simulations.

**Simulating Game 1** Running a large number of simulations to estimate the expected outcomes and analyze the distribution of prizes

```python
# Creating a DataFrame to store simulation results for Game 1
results_game01 = pd.DataFrame({
    "Letter1":[],
    "Letter2":[],
    "Letter3":[],
    "Number1":[],
    "Number2":[],
    "Prize":[]
})

# Defining the number of simulations to approximate the distribution of outcomes
num_plays = 1000000

# Running the simulation multiple times
for i in range(num_plays):
    
    # Executing the game function to generate outcome and prize
    outcome, prize = game01()
    
    # Storing the result of each simulation in a temporary DataFrame
    new_results = pd.DataFrame({
        "Letter1":[outcome[0]],
        "Letter2":[outcome[1]],
        "Letter3":[outcome[2]],
        "Number1":[outcome[3]],
        "Number2":[outcome[4]],
        "Prize":[prize]
    })
    
    # Appending the new result to the main DataFrame
    results_game01 = pd.concat([results_game01, new_results], ignore_index=True)

# Calculating measures of central tendency for the prize distribution
print("Mean:", st.mean(results_game01["Prize"]))
print("Median:", st.median(results_game01["Prize"]))
print("Mode:", st.mode(results_game01["Prize"]))
```

The mean indicates a small expected payout, while both the median and mode show that the most common outcome is zero. This suggests that the game is highly profitable for the club, as the vast majority of simulations result in no prize being awarded, minimizing overall payout costs.

**Simulating Game 2** multiple times to analyze prize distribution in a case-sensitive scenario.

```python
# Creating a DataFrame to store simulation results for Game 2
results_game2 = pd.DataFrame({
    "Letter1":[],
    "Letter2":[],
    "Letter3":[],
    "Prize":[]
})

# Defining the number of simulations to approximate the distribution of outcomes
num_plays = 1000000

# Running multiple simulations of Game 2
for i in range(num_plays):
    
    # Executing the game function to generate outcome and prize
    outcome, prize = game2()
    
    # Storing the result of each simulation in a temporary DataFrame
    new_results = pd.DataFrame({
        "Letter1":[outcome[0]],
        "Letter2":[outcome[1]],
        "Letter3":[outcome[2]],
        "Prize":[prize]
    })
    
    # Appending the new result to the main DataFrame
    results_game2 = pd.concat([results_game2, new_results], ignore_index=True)

# Calculating measures of central tendency for Game 2
print("Mean:", st.mean(results_game2["Prize"]))
print("Median:", st.median(results_game2["Prize"]))
print("Mode:", st.mode(results_game2["Prize"]))
```

Based on the calculated values of the **mean, mode, and median**, it can be observed that in Game 1 there is a higher probability that the club will spend money on prize payouts over time. In contrast, Game 2, which presents a mean equal to zero, indicates that the probability of the club paying out prizes is lower over time compared to Game 1.

# Task 4) Calculate the measures of dispersion for each game, based off an appropriate number of simulations.

Analyzing the variability of prize outcomes to understand the level of risk in **Game 1**

```python
# Calculating measures of dispersion for Game 1

# The interquartile range (IQR)

Q1_game1 =  results_game01['Prize'].quantile(0.25)
Q3_game1 =  results_game01['Prize'].quantile(0.75)

IQR_game1 = Q3_game1 - Q1_game1

# Standard deviation indicates the average distance from the mean
std_dev_game01 = results_game01["Prize"].std()

# Range shows the difference between the maximum and minimum prize values
range_value_game01 = results_game01["Prize"].max() - results_game01["Prize"].min()

# Printing the results
print("Game 1 IQR:", IQR_game1)
print("Standard Deviation Game 1:", std_dev_game01)
print("Range Game 1:", range_value_game01)
```

The dispersion measures indicate that most prize values are concentrated around zero, suggesting low variability in outcomes.

Evaluating the variability of outcomes in **Game 2** to assess risk and payout dispersion

```python
# Calculating measures of dispersion for Game 2

# The interquartile range (IQR)
Q1_game2 =  results_game2['Prize'].quantile(0.25)
Q3_game2 =  results_game2['Prize'].quantile(0.75)

IQR_game2 = Q3_game2 - Q1_game2

# Standard deviation indicates the average distance from the mean
std_dev_game02 = results_game2["Prize"].std()

# Range shows the difference between the maximum and minimum prize values
range_value_game02 = results_game2["Prize"].max() - results_game2["Prize"].min()

# Printing the results
print("Game 2 IQR:", IQR_game2)
print("Standard Deviation Game 2:", std_dev_game02)
print("Range Game 2:", range_value_game02)
```

# Task 5) Discuss, in terms of those calculated statistics, which game you would recommend the club to adopt.

```python
ticket_price = 5

summary = pd.DataFrame({
    "Metric": [
        "Mean",
        "Median",
        "Mode",
        "IQR",
        "Standard Deviation",
        "Range",
        "Club Profit"
    ],
    "Game 1": [
        results_game01["Prize"].mean(),
        results_game01["Prize"].median(),
        results_game01["Prize"].mode()[0],
        IQR_game1,
        results_game01["Prize"].std(),
        results_game01["Prize"].max() - results_game01["Prize"].min(),
        ticket_price - expected_value_game01
    ],
    "Game 2": [
        results_game2["Prize"].mean(),
        results_game2["Prize"].median(),
        results_game2["Prize"].mode()[0],
        IQR_game2,
        results_game2["Prize"].std(),
        results_game2["Prize"].max() - results_game2["Prize"].min(),
        ticket_price - expected_value_game02
    ]
})

summary
```

To evaluate the proposed games, the statistical measures calculated above are used, including expected value, measures of central tendency, and dispersion. These calculations aim to determine which game is more advantageous for the club from a financial perspective.

Firstly, the expected value is a crucial metric, as it represents the average profit the club expects to obtain per game over the long term. Based on the results, Game 1 has an expected value of approximately €4.90, while Game 2 has an expected value of approximately €1.44. This indicates that Game 1 is more profitable and provides a higher average return to the club when compared to Game 2.

In addition to the expected value, the mean values also provide important insights. Game 1 has a mean of approximately 0.285, while Game 2 has a mean equal to 0. This suggests that in Game 1 there is a higher probability of prize payouts compared to Game 2. However, even with this higher probability of payouts, Game 1 still remains the better option. This is because, when analysing the median and mode, both games present values equal to 0. This indicates that, in the majority of cases, players do not win any prize in either game.

Therefore, even though Game 1 includes occasional payouts, these payments are relatively small compared to the potential payout in Game 2. As a result, the financial risk associated with Game 1 is lower. In other words, if the club needs to pay a prize in Game 1, the amount is manageable and does not significantly impact overall profitability.

Furthermore, the interquartile range (IQR) is equal to 0 for both games. This supports the previous conclusion that most outcomes result in no prize being awarded. However, it is important to note that the IQR does not capture extreme values, such as outliers or rare events. For this reason, it is not sufficient on its own to evaluate the financial risk of the games.

To better understand the variability, it is necessary to consider additional dispersion measures. Game 1 presents a standard deviation of approximately 2.32 and a range of 25. This indicates a moderate level of variability, meaning that while some prizes are paid, they are not frequent and are relatively small. This makes Game 1 more predictable and financially manageable for the club.

On the other hand, Game 2 has a standard deviation and range equal to 0, which reflects the absence of winning events during the simulations. However, this does not mean that the game is risk-free. The absence of payouts is due to the extremely low probability of winning, rather than the absence of prizes. If a player were to win the €500,000 prize, the club would face a substantial financial loss. This represents a high-impact, low-probability risk that is not captured in the simulation results.

Therefore, while Game 2 may initially appear more advantageous due to the lack of observed payouts, this interpretation is misleading. The potential financial loss associated with a single winning event makes this game significantly riskier.

In contrast, Game 1 offers a more balanced financial structure, with smaller and more controlled payouts. This results in a more stable and predictable model for the club, where the risk is distributed across smaller amounts rather than concentrated in a single extreme event.

In conclusion, although Game 2 appears to be risk-free in the simulations, it carries a significant financial risk due to the high-value prize. Game 1, on the other hand, provides a higher expected value, controlled variability, and a more sustainable profit structure. Therefore, Game 1 is the recommended option for the club, as it offers a better balance between profitability and financial risk.
