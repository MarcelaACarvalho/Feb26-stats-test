## Task 1) Calculate the expected value for the club for both games.

```python
import random
import string
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
import statistics as st
```

```python
## GAME 1

# The player pays €5 and picks 3 letters (all lower case English letters) and 2 digits (numbers 0-9 inclusive). They win as follows:

    # What is correct            Prize
    # 3 letters, both numbers	  €100
    # Two letters, both numbers	  €50
    # One letter, both numbers	  €25
    # No letter, both numbers	  €10

# Matched x Approximate Odds

all_letters = 1/26 * 1/26 * 1/26 * 1/10 * 1/10
print(all_letters)
two_letters = 1/26 * 1/26 * 25/26 * 1/10 * 1/10
print(two_letters)
one_letters = 1/26 * 25/25 * 25/26 * 1/10 * 1/10
print(one_letters)
no_letters = 25/26 * 25/25 * 25/26 * 1/10 * 1/10
print(no_letters)
```

```python
jackpot = 100 #This is the maximum amount a person can win in this game.
win_data = [
    [jackpot, all_letters],
    [50, two_letters],
    [25, one_letters],
    [10, no_letters]
]

expected_value_game01 = 0
for wins in win_data:
    expected_value_game01 += wins[0] * wins[1]
    print(f"Ex: {wins[0] * wins[1]}")
print("expected_value_player_game1:", expected_value_game01)
print("expected_value_club_game1:",5 - expected_value_game01) # The 5 represents the amount paid for each play.
```

```python
# Expected value means how much the player or club will "receive" on average for each play.
# In this case the player will "receive" 0.10 euros and the club will "receive" 4.90 euros.
# In this game, it's always much more advantageous for the club than for the player.
```

```python
# On average, you lose about €4.9 per game and win €0.10
# Imagine playing 10,000 times:

# You would spend:
# → €5 × 10,000 = €50,000
# You would win (on average):
# → €0.12 × 10,000 = €1,200

# For the club, the thinking is exactly the opposite:
# It earns €4.90 for each ticket
```

```python
## GAME 02

#The player pays €5 and picks 3 letters (all lower case AND upper case English letters, repetition allowed). 
#If they guess all correctly, they win €500,000
```

```python
# Matched x Approximate Odds

letters = 1/52 * 1/52 * 1/52
print(letters)

jackpot = 500_000
expected_value_game02 = jackpot * letters

print("expected_value_player:", expected_value_game02)
print("expected_value_club:", 5 - expected_value_game02)
```

```python
# On average, you lose about €1.44 per game and win €3.55
# Imagine playing 10,000 times:

# You would spend:
# → €5 × 10,000 = €50,000
# You would win (on average):
# → €3.55 × 10,000 = €3,550

# For the club, the thinking is exactly the opposite:
# It wins €1.44 for each tickets

# In this game, the player has a higher average gain than in the first game, but it is still more advantageous for the club than for the player.
```

# Task 2) Create simulations of each game in Python

```python
## simulations`s Game 01

import random
import string
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
import statistics as st
```

```python
# Simulating a random event using a probability value between 0 and 1
# This value will determine whether the letter is changed or kept

def spin_letter():
    outcome = random.random()
    
    if outcome < 1/26:
        return "a"
    elif outcome < 2/26:
        return "b"
    elif outcome < 3/26:
        return "c"
    elif outcome < 4/26:
        return "d"
    elif outcome < 5/26:
        return "e"
    elif outcome < 6/26:
        return "f"
    elif outcome < 7/26:
        return "g"
    elif outcome < 8/26:
        return "h"
    elif outcome < 9/26:
        return "i"
    elif outcome < 10/26:
        return "j"
    elif outcome < 11/26:
        return "k"
    elif outcome < 12/26:
        return "l"
    elif outcome < 13/26:
        return "m"
    elif outcome < 14/26:
        return "n"
    elif outcome < 15/26:
        return "o"
    elif outcome < 16/26:
        return "p"
    elif outcome < 17/26:
        return "q"
    elif outcome < 18/26:
        return "r"
    elif outcome < 19/26:
        return "s"
    elif outcome < 20/26:
        return "t"
    elif outcome < 21/26:
        return "u"
    elif outcome < 22/26:
        return "v"
    elif outcome < 23/26:
        return "w"
    elif outcome < 24/26:
        return "x"
    elif outcome < 25/26:
        return "y"
    else:
        return "z"
```

```python
# Executing the spin_letter function to simulate a random letter outcome
spin_letter()
```

```python
# This value will determine whether the number is changed or kept

def spin_number():
    outcome = random.random()
    
    if outcome < 1/10:
        return 0
    elif outcome < 2/10:
        return 1
    elif outcome < 3/10:
        return 2
    elif outcome < 4/10:
        return 3
    elif outcome < 5/10:
        return 4
    elif outcome < 6/10:
        return 5
    elif outcome < 7/10:
        return 6
    elif outcome < 8/10:
        return 7
    elif outcome < 9/10:
        return 8
    else:
        return 9
```

```python
# Executing the spin_letter function to simulate a random number outcome
spin_number()
```

```python
# Game Simulation 1: simulating the lottery game to evaluate outcomes and prize distribution

# Defining the player's chosen letters and numbers
player_letters = ["a", "b", "c"]
player_numbers = [1, 2]

def game01():
    
    # Generating random letters
    letter1 = spin_letter()
    letter2 = spin_letter()
    letter3 = spin_letter()
    
    # Generating random numbers
    number1 = spin_number()
    number2 = spin_number()
    
    # Storing the generated results
    outcome_letters = [letter1, letter2, letter3]
    outcome_numbers = [number1, number2]

    # Initializing counters for correct matches
    correct_letters = 0
    correct_numbers = 0

    # Counting how many player letters match the outcome
    for letter in player_letters:
        if letter in outcome_letters:
            correct_letters += 1

    # Counting how many player numbers match the outcome
    for number in player_numbers:
        if number in outcome_numbers:
            correct_numbers += 1

    # Assigning prize based on matching rules
    if correct_numbers == 2 and correct_letters == 3:
        prize = 100
    elif correct_numbers == 2 and correct_letters == 2:
        prize = 50
    elif correct_numbers == 2 and correct_letters == 1:
        prize = 25
    elif correct_numbers == 2 and correct_letters == 0:
        prize = 10
    else:
        prize = 0

    # Combining letters and numbers into a single outcome
    outcome = outcome_letters + outcome_numbers

    # Returning the final result (outcome and prize)
    return outcome, prize
```

```python
# Running the first game simulation
game01()
```

```python
# Defining the player's chosen letters (case-sensitive)
player_letters = ["a", "B", "c"]

# Simulating Game 2 with case-sensitive letter matching
def game2():
    
    # Generating three random letters using the spin function
    letter1 = spin_letter2()
    letter2 = spin_letter2()
    letter3 = spin_letter2()

    # Storing the generated outcome as a list
    outcome = [letter1, letter2, letter3]

    # Checking if the generated outcome matches the player's letters exactly
    if outcome == player_letters:
        prize = 500000  # Jackpot prize
        
    else:
        prize = 0  # No prize

    # Returning the result of the simulation (outcome and prize)
    return outcome, prize
```

In this version of the game, uppercase and lowercase letters are treated as different values,
making the probability of winning significantly lower

## Task 3) Calculate the measures of central tendency for each game, based off an appropriate number of simulations.

Running a large number of simulations to estimate the expected outcomes and analyze the distribution of prizes

```python
# Creating a DataFrame to store simulation results for Game 1
results_game01 = pd.DataFrame({
    "Letter1":[],
    "Letter2":[],
    "Letter3":[],
    "Number1":[],
    "Number2":[],
    "Prize":[]
})

# Defining the number of simulations to approximate the distribution of outcomes
num_plays = 100000

# Running the simulation multiple times
for i in range(num_plays):
    
    # Executing the game function to generate outcome and prize
    outcome, prize = game01()
    
    # Storing the result of each simulation in a temporary DataFrame
    new_results = pd.DataFrame({
        "Letter1":[outcome[0]],
        "Letter2":[outcome[1]],
        "Letter3":[outcome[2]],
        "Number1":[outcome[3]],
        "Number2":[outcome[4]],
        "Prize":[prize]
    })
    
    # Appending the new result to the main DataFrame
    results_game01 = pd.concat([results_game01, new_results], ignore_index=True)

# Calculating measures of central tendency for the prize distribution
print("Mean:", st.mean(results_game01["Prize"]))
print("Median:", st.median(results_game01["Prize"]))
print("Mode:", st.mode(results_game01["Prize"]))
```

The mean suggests a small average return, but the median and mode show that the most common outcome is zero. 
This indicates that the game is not favourable to the player.

Simulating Game 2 multiple times to analyze prize distribution in a case-sensitive scenario.

```python
# Creating a DataFrame to store simulation results for Game 2
results_game2 = pd.DataFrame({
    "Letter1":[],
    "Letter2":[],
    "Letter3":[],
    "Prize":[]
})

# Defining the number of simulations to approximate the distribution of outcomes
num_plays = 100000

# Running multiple simulations of Game 2
for i in range(num_plays):
    
    # Executing the game function to generate outcome and prize
    outcome, prize = game2()
    
    # Storing the result of each simulation in a temporary DataFrame
    new_results = pd.DataFrame({
        "Letter1":[outcome[0]],
        "Letter2":[outcome[1]],
        "Letter3":[outcome[2]],
        "Prize":[prize]
    })
    
    # Appending the new result to the main DataFrame
    results_game2 = pd.concat([results_game2, new_results], ignore_index=True)

# Calculating measures of central tendency for Game 2
print("Mean:", st.mean(results_game2["Prize"]))
print("Median:", st.median(results_game2["Prize"]))
print("Mode:", st.mode(results_game2["Prize"]))
```

Based on the mean, mode, and median values, Game 1 has a positive average prize, indicating that the club will incur some costs over time due to prize payments. In contrast, Game 2 has an average of zero, meaning that no prize is actually awarded. Therefore, Game 2 is more advantageous for the club as it minimizes payment costs and maximizes profitability.
