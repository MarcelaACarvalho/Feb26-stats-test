## Task 1) Calculate the expected value for the club for both games.

```python
import random
import string
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
import statistics as st
```

```python
## GAME 1

# The player pays €5 and picks 3 letters (all lower case English letters) and 2 digits (numbers 0-9 inclusive). They win as follows:

    # What is correct            Prize
    # 3 letters, both numbers	  €100
    # Two letters, both numbers	  €50
    # One letter, both numbers	  €25
    # No letter, both numbers	  €10

# Matched x Approximate Odds

all_letters = 1/26 * 1/26 * 1/26 * 1/10 * 1/10
print(all_letters)
two_letters = 1/26 * 1/26 * 25/26 * 1/10 * 1/10
print(two_letters)
one_letters = 1/26 * 25/25 * 25/26 * 1/10 * 1/10
print(one_letters)
no_letters = 25/26 * 25/25 * 25/26 * 1/10 * 1/10
print(no_letters)
```

```python
jackpot = 100 #This is the maximum amount a person can win in this game.
win_data = [
    [jackpot, all_letters],
    [50, two_letters],
    [25, one_letters],
    [10, no_letters]
]

expected_value_game01 = 0
for wins in win_data:
    expected_value_game01 += wins[0] * wins[1]
    print(f"Ex: {wins[0] * wins[1]}")
print("expected_value_player_game1:", expected_value_game01)
print("expected_value_club_game1:",5 - expected_value_game01) # The 5 represents the amount paid for each play.
```

```python
# Expected value means how much the player or club will "receive" on average for each play.
# In this case the player will "receive" 0.10 euros and the club will "receive" 4.90 euros.
# In this game, it's always much more advantageous for the club than for the player.
```

```python
# On average, you lose about €4.9 per game and win €0.10
# Imagine playing 10,000 times:

# You would spend:
# → €5 × 10,000 = €50,000
# You would win (on average):
# → €0.12 × 10,000 = €1,200

# For the club, the thinking is exactly the opposite:
# It earns €4.90 for each ticket
```

```python
## GAME 02

#The player pays €5 and picks 3 letters (all lower case AND upper case English letters, repetition allowed). 
#If they guess all correctly, they win €500,000
```

```python
# Matched x Approximate Odds

letters = 1/52 * 1/52 * 1/52
print(letters)

jackpot = 500_000
expected_value_game02 = jackpot * letters

print("expected_value_player:", expected_value_game02)
print("expected_value_club:", 5 - expected_value_game02)
```

```python
# On average, you lose about €1.44 per game and win €3.55
# Imagine playing 10,000 times:

# You would spend:
# → €5 × 10,000 = €50,000
# You would win (on average):
# → €3.55 × 10,000 = €3,550

# For the club, the thinking is exactly the opposite:
# It wins €1.44 for each tickets

# In this game, the player has a higher average gain than in the first game, but it is still more advantageous for the club than for the player.
```
